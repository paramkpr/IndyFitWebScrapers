The general process is to query the public API to find the links of all recipes. Then we will do some URL manipulation to get the 'print' recipe pages so that the actual scraping is faster, easier, and cleaner. However, there are some data points that are not on the 'print recipe' page, for those we will query the actual page and retrive them. This seems like double the work but it should be worth it.

### Getting url or slugs of all recipes

```python
import requests 
import json
from bs4 import BeautifulSoup
from pprint import pprint


url = "https://www.sanjeevkapoor.com/AllRecipes.aspx/GetDataFromServer"

payload = "{\"divIndex\": 17984}"
print(payload)
headers= {'cookie': 'ASP.NET_SessionId=us35q0aki2vk0ffmtpe2tvm0; Path=/; Domain=www.sanjeevkapoor.com; HttpOnly;',
         'Content-Type': 'application/json'}

r = requests.request("POST", url, headers=headers, data = payload)

html = r.json()['d']

soup = BeautifulSoup(html, 'html.parser')
urls = ['https://www.sanjeevkapoor.com' + i['href'][2:] for i in soup.findAll('a')]
print(urls, len(urls))

```

```python
url = "https://www.sanjeevkapoor.com/AllRecipes.aspx/GetDataFromServer"
headers= {'cookie': 'ASP.NET_SessionId=us35q0aki2vk0ffmtpe2tvm0; Path=/; Domain=www.sanjeevkapoor.com; HttpOnly;',
         'Content-Type': 'application/json'}


with open('links.txt', 'w') as f:
    for i in range(17984, 2384, -6):
        payload = """{"divIndex": """ + str(i) + """}"""
        r = requests.request("POST", url, headers=headers, data = payload)
        html = r.json()['d']

        soup = BeautifulSoup(html, 'html.parser')
        for j in soup.findAll('a'):
            f.write('https://www.sanjeevkapoor.com' + j['href'][2:] + '\n')
        print('got recipe batch ' + str(i))
        
```

### Getting Data from _Print Recipe_ Page

```python
url = 'https://www.sanjeevkapoor.com/recipe/printrecipe.aspx?recipe_name=Pulled%20Chicken%20Uttapam%20-%20SK%20Khazana'
html = requests.get(url).content
soup = BeautifulSoup(html, 'html.parser')


def get_ingredients(soup):
    li = soup.find('div', {'class': 'ingrdntdiv'}).ul
    list_items = li.findAll('li')
    x = []
    for i in list_items:
        x.append(i.text.strip())
    return x
    

r = {'title': soup.find('h3', {'id': 'headingh1'}).text.split('- SK')[0].strip(),
     'description': soup.find('div', {'class': 'shortdisc', 'itemprop': 'description'}).text.strip(),
     'prep_time': soup.find('meta', {'itemprop': 'prepTime'}).parent.text.split(':')[-1].strip(),
     'cook_time': soup.find('meta', {'itemprop': 'cookTime'}).parent.text.split(':')[-1].strip(),
     'servings': soup.find('span', {'itemprop': 'recipeYield'}).text,
     'ingredients': get_ingredients(soup),
     'method': [i.text.strip() for i in soup.findAll('span', {'itemprop': 'recipeInstructions'})],
    }
pprint(r)
```
