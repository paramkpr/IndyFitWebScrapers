```python
from requests import get
from bs4 import BeautifulSoup
# from panda import DataFrame
from pprint import pprint


def is_recipe(soup):
    try:
        soup.find('a', {'class': 'wprm-recipe-jump'}).span
    except AttributeError:
        return False
    return True

def get_links():
    r = get('https://www.indianhealthyrecipes.com/recipe-index/', headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    links =[]
    for link in soup.findAll('a', {'class': '_self cvplbd'}):
        s = BeautifulSoup(get(link['href'], 'html.parser').content)
        if is_recipe(s):
            links.append(link['href'])
            print('got link ' + link['href'])
    
    return links[2:]

links = get_links()

with open('links.txt', 'w') as f:
    for ln in links:
        f.write(ln + '\n')
        
```
